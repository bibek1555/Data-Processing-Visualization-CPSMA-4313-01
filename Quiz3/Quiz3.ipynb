```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/iris.csv"
iris = pd.read_csv(url)
```

```python
iris.head()
```

```python
iris.info()
```

```python
iris.describe()
```

```python
iris['Class'].value_counts().plot(kind='bar')
plt.title("Count of Iris Classes")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(iris['SepalLength'])
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
#Visualizations of your favorite variable, mine was SepalLength

plt.hist(iris['SepalLength'])
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.boxplot(iris['SepalLength'])
plt.title("Box Plot of Sepal Length")
plt.ylabel("Sepal Length")
plt.show()
```

```python
plt.violinplot(iris['SepalLength'])
plt.title("Violin Plot of Sepal Length")
plt.ylabel("Sepal Length")
plt.show()
```
