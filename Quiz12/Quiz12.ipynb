In this notebook, I will:

- Gather an HTML table from a website using Python
- Use requests and BeautifulSoup to read the webpage
- Load the table into a pandas DataFrame
- Export the table as a CSV file

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from io import StringIO
```

```python

headers = {'User-Agent': 'Mozilla/5.0'}
url = "https://en.wikipedia.org/wiki/The_Simpsons"

r = requests.get(url, headers=headers)
html_contents = r.text
html_soup = BeautifulSoup(html_contents, "html.parser")

tables = html_soup.find_all("table")

df = pd.read_html(StringIO(str(tables)))[2]
df.columns = df.columns.droplevel(0).droplevel(0)
df.head()
```

```python
df.to_csv("simpsons_table.csv", index=False)
print("Table exported successfully as simpsons_table.csv")
```
