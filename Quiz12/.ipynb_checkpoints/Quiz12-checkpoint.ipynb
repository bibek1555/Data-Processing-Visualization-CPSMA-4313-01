In this notebook, I will:

- Gather an HTML table from the House (TV series) Wikipedia page using Python
- Use requests and BeautifulSoup to read the webpage
- Load the table into a pandas DataFrame
- Export the table as a CSV file

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from io import StringIO
```

```python
headers = {'User-Agent': 'Mozilla/5.0'}
url = "https://en.wikipedia.org/wiki/House_(TV_series)"

r = requests.get(url, headers=headers)
html_contents = r.text
html_soup = BeautifulSoup(html_contents, "html.parser")

tables = html_soup.find_all("table")
len(tables)
```

```python
df = pd.read_html(StringIO(str(tables[2])))[0]
print(df.columns)
print(df.shape)
```

```python
df = pd.read_html(StringIO(str(tables[2])))[0]
df.columns = df.columns.get_level_values(1)
df = df.drop(columns=["Episodes.1"])
df.head()
```

```python
df.to_csv("house_table.csv", index=False)
print("Table exported successfully as house_table.csv")
```
