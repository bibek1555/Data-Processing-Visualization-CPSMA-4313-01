## Tables Used

- `bigquery-public-data.thelook_ecommerce.orders`
- `bigquery-public-data.thelook_ecommerce.users`

These tables can be joined using:

- `orders.user_id = users.id`

This relationship is appropriate because each order is associated with a user.

## Preview of the Users Table

```python
%%bigquery --project quiz11-bigquery-sql-joins

SELECT *
FROM `bigquery-public-data.thelook_ecommerce.users`
LIMIT 10;
```

## INNER JOIN

An INNER JOIN returns only the rows that have matching values in both tables.

In this example, only orders with a matching user record are returned.

```python
%%bigquery --project quiz11-bigquery-sql-joins

SELECT
  o.order_id,
  o.user_id,
  o.status,
  u.first_name,
  u.last_name,
  u.email
FROM `bigquery-public-data.thelook_ecommerce.orders` AS o
INNER JOIN `bigquery-public-data.thelook_ecommerce.users` AS u
  ON o.user_id = u.id
LIMIT 20;
```

## LEFT JOIN

A LEFT JOIN returns all rows from the left table and matching rows from the right table.

If no match exists in the right table, the result shows `NULL` for the right-side columns.

Here, all rows from `orders` are kept.

```python
%%bigquery --project quiz11-bigquery-sql-joins

SELECT
  o.order_id,
  o.user_id,
  o.status,
  u.first_name,
  u.last_name,
  u.email
FROM `bigquery-public-data.thelook_ecommerce.orders` AS o
LEFT JOIN `bigquery-public-data.thelook_ecommerce.users` AS u
  ON o.user_id = u.id
LIMIT 20;
```

## RIGHT JOIN

A RIGHT JOIN returns all rows from the right table and matching rows from the left table.

If no match exists in the left table, the result shows `NULL` for the left-side columns.

Here, all rows from `users` are kept.

```python
%%bigquery --project quiz11-bigquery-sql-joins

SELECT
  o.order_id,
  o.user_id,
  o.status,
  u.id AS user_id,
  u.first_name,
  u.last_name,
  u.email
FROM `bigquery-public-data.thelook_ecommerce.orders` AS o
RIGHT JOIN `bigquery-public-data.thelook_ecommerce.users` AS u
  ON o.user_id = u.id
LIMIT 20;
```

## FULL OUTER JOIN

A FULL OUTER JOIN returns all rows from both tables.

Matching rows are combined. Non-matching rows from either side are also included, with `NULL` values where data is missing.

```python
%%bigquery --project quiz11-bigquery-sql-joins

SELECT
  o.order_id,
  o.user_id,
  o.status,
  u.id AS user_id,
  u.first_name,
  u.last_name,
  u.email
FROM `bigquery-public-data.thelook_ecommerce.orders` AS o
FULL OUTER JOIN `bigquery-public-data.thelook_ecommerce.users` AS u
  ON o.user_id = u.id
LIMIT 20;
```

## Comparing row counts for all four joins

This query summarizes the number of rows returned by each join type.

```python
%%bigquery --project quiz11-bigquery-sql-joins

SELECT 'INNER JOIN' AS join_type, COUNT(*) AS row_count
FROM `bigquery-public-data.thelook_ecommerce.orders` AS o
INNER JOIN `bigquery-public-data.thelook_ecommerce.users` AS u
  ON o.user_id = u.id

UNION ALL

SELECT 'LEFT JOIN' AS join_type, COUNT(*) AS row_count
FROM `bigquery-public-data.thelook_ecommerce.orders` AS o
LEFT JOIN `bigquery-public-data.thelook_ecommerce.users` AS u
  ON o.user_id = u.id

UNION ALL

SELECT 'RIGHT JOIN' AS join_type, COUNT(*) AS row_count
FROM `bigquery-public-data.thelook_ecommerce.orders` AS o
RIGHT JOIN `bigquery-public-data.thelook_ecommerce.users` AS u
  ON o.user_id = u.id

UNION ALL

SELECT 'FULL OUTER JOIN' AS join_type, COUNT(*) AS row_count
FROM `bigquery-public-data.thelook_ecommerce.orders` AS o
FULL OUTER JOIN `bigquery-public-data.thelook_ecommerce.users` AS u
  ON o.user_id = u.id;
```
