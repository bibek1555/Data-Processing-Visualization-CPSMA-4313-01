In this notebook, we will:

- Load the Iris dataset
- Plot Sepal Length vs Sepal Width
- Use different colors for each species
- Explain how colors are assigned


```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df.head()
```

## Understanding the Species Column

The species column contains numeric values:

- 0 → Setosa
- 1 → Versicolor
- 2 → Virginica

We will assign a different color to each species manually.

```python
colors = ['red', 'green', 'blue']

for i in range(3):
    plt.scatter(
        df[df['species'] == i]['sepal length (cm)'],
        df[df['species'] == i]['sepal width (cm)'],
        color=colors[i],
        label=iris.target_names[i]
    )

plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Sepal Length vs Width by Species")
plt.legend()
plt.show()
```

## How Colors Are Assigned

We created a list of colors:

    colors = ['red', 'green', 'blue']

Then we looped through species (0, 1, 2).

For each species:
- We filtered the dataframe
- Plotted only that species
- Assigned one specific color

This ensures that each class has a distinct color.

## Why It May Not Have Worked in Class

If we try:

    plt.scatter(x, y, c=df['species'])

Matplotlib treats species as continuous values, not categories.
This creates a gradient instead of 3 distinct colors.


```python
plt.scatter(
    df['sepal length (cm)'],
    df['sepal width (cm)'],
    c=df['species'],
    cmap='viridis'
)

plt.colorbar()
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Using Colormap Instead")
plt.show()
```
