```python
#Task 1 : Create a bar graph of the maximum ‘price’ by ‘neighbourhood_group’. Include the ‘price’ in your graph

import pandas as pd
import matplotlib.pyplot as plt

# Load data
url = "https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/AB_NYC_2019.csv"
df = pd.read_csv(url)

# Aggregate: maximum price by neighbourhood_group
max_price = df.groupby("neighbourhood_group")["price"].max()

# Bar chart
plt.figure(figsize=(8, 5))
plt.bar(max_price.index, max_price.values)

plt.xlabel("Neighbourhood Group")
plt.ylabel("Maximum Price")
plt.title("Maximum Airbnb Price by Neighbourhood Group")
plt.ylim(0, max_price.max() * 1.05)

plt.show()
```

```python
#Task 2:Create a multiple bar graph with ‘neighbourhood_group’ and ‘room_type’ by looking at the average ‘price’.


import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/AB_NYC_2019.csv"
df = pd.read_csv(url)

avg_price = (
    df.groupby(["neighbourhood_group", "room_type"])["price"]
      .mean()
      .unstack()
)

avg_price.plot(kind="bar", figsize=(10, 6))

plt.xlabel("Neighbourhood Group")
plt.ylabel("Average Price")
plt.title("Average Airbnb Price by Neighbourhood Group and Room Type")

# Bar charts must start at zero
plt.ylim(0, avg_price.max().max() * 1.1)

plt.legend(title="Room Type")
plt.show()
```
