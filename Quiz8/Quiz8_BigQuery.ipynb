In this notebook, we will:

- Create and use a Google Cloud project to run BigQuery queries  
- Navigate to the dataset `bigquery-public-data.austin_bikeshare.bikeshare_trips`  
- Compute how many entries are in the dataset  
- Compute the longest trip using `duration_minutes`  
- Compute the average trip time using `duration_minutes`  

```python
from google.colab import auth
auth.authenticate_user()
print("Authenticated")
```

```python
%%bigquery --project quiz8-bigquery
SELECT COUNT(*) AS total_rows
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`;
```

```python
%%bigquery --project quiz8-bigquery
SELECT MAX(duration_minutes) AS longest_trip_minutes
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`;
```

```python
%%bigquery --project quiz8-bigquery
SELECT ROUND(AVG(duration_minutes), 2) AS avg_trip_minutes
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`;
```
