## Loading the Dataset

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/AB_NYC_2019.csv")

df.head()
```

## Checking Missing Values

```python
df.isnull().sum()
```

### The output shows how many missing values exist in each column. This helps identify which columns need cleaning before further analysis.

## Total Number of Missing Values

```python
df.isnull().sum().sum()
```

## Imputing Missing Values

```python
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
```

### The `reviews_per_month` column was filled with 0 because missing values likely represent listings that have not received any reviews yet.

```python
df['last_review'] = df['last_review'].fillna("No Review")
```

### The `last_review` column was filled with `"No Review"` because listings with no review history do not have a valid last review date.

## Verifying the Changes

```python
df[['reviews_per_month', 'last_review']].isnull().sum()
```

### After imputation, the selected columns no longer contain missing values. This completes the missing data cleaning requirement for the quiz.
